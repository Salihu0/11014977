# Create a figure and axes

```python
fig, axs = plt.subplots(ncols=5, nrows=3, figsize=(15, 10))

# Plot KDE plots
axs[0, 0].plot(data['fixed acidity'])
axs[0, 1].plot(data['volatile acidity'])
axs[0, 2].plot(data['citric acid'])
axs[0, 3].plot(data['residual sugar'])
axs[0, 4].plot(data['chlorides'])
axs[1, 0].plot(data['free sulfur dioxide'])
axs[1, 1].plot(data['total sulfur dioxide'])
axs[1, 2].plot(data['density'])
axs[1, 3].plot(data['pH'])
axs[1, 4].plot(data['sulphates'])
axs[2, 0].plot(data['alcohol'])
axs[2, 1].plot(data['quality'])

# Set titles and labels
axs[0, 0].set_title('Fixed Acidity')
axs[0, 1].set_title('Volatile Acidity')
axs[0, 2].set_title('Citric Acid')
axs[0, 3].set_title('Residual Sugar')
axs[0, 4].set_title('Chlorides')
axs[1, 0].set_title('Free Sulfur Dioxide')
axs[1, 1].set_title('Total Sulfur Dioxide')
axs[1, 2].set_title('Density')
axs[1, 3].set_title('pH')
axs[1, 4].set_title('Sulphates')
axs[2, 0].set_title('Alcohol')
axs[2, 1].set_title('Quality')

# Adjust spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Pie chart of a categorical column
plt.figure(figsize=(8,6)) # set the figure size
quality_counts = data['quality'].value_counts()
plt.pie(quality_counts.values, labels=quality_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart of Quality')
plt.axis('equal')
plt.show()
```

##### REPORT ON THE PIE CHART 
- The quality of 5.0 and 6.0 have a high percentage that means there are more red wine with these qualities i.e 5.0 and 6.0
-  And also the quality of 3.0 and 8.0 are not that many and the 7.0 quality is in the middle. The 4.0 quality is the next highest after the 7.0 quality

# Boxplot of a numerical column

```python

plt.figure(figsize=(8,6)) # set the figure size
plt.boxplot(x=data['alcohol'])
plt.title('Boxplot of Alcohol')
plt.ylabel('Alcohol')
plt.show()
```

### REPORT ON BOXPLOT OF ALCOHOL 
- The alcohol content at the median level is approximately 10%.  
- The box plot indicates a skewness towards the lower end.  
- This suggests that there is a clustering of wines with lower alcohol content.

## Distribution of residual sugar in red wine

```python
sugar_distribution = data['residual sugar'].value_counts()

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(sugar_distribution.index, sugar_distribution, color=['blue', 'red', 'grey', 'yellow', 'green'])
ax.set_title('Distribution of Residual Sugar in Red Wine')
ax.set_xlabel('Residual Sugar (g/L)')
ax.set_ylabel('Frequency')

plt.show()
```

## Analysis Of The Bar Graph Above  
* The graph indicates a large number of entries for residual sugar in the red wine dataset. 
* ⁠ To improve readability, the data has been grouped into three labels:   
* ⁠- 'Low' for residual sugars ranging from 1.0 to 5.9.   
* ⁠- 'Medium' for residual sugars ranging from 6.0 to 9.9.   
* ⁠- 'High' for residual sugars ranging from 9.9 to 13.9. 
* ⁠This categorization simplifies the interpretation of the graph.

### Bar Graph For The  Distribution of Alcohol content distribution

```python
alcohol_content = data['alcohol']

plt.figure(figsize=(10, 6))
plt.hist(alcohol_content, bins=15, color='blue', edgecolor='black')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.title('Alcohol Content Distribution in Red Wine')

plt.show()
```

### REPORT ON THR GRAPH ABOVE 
- Based on the provided graph, we can observe the distribution of alcohol content in red wine entries.  
- There is a significant frequency of red wines with alcohol content between 9 and 10.  
- The frequency of red wines with alcohol content of 11 is at the median level.  
- However, the frequency of red wines with alcohol content between 12 and 15 is quite low.  
- Therefore, there are not many red wines with alcohol content between 12 and 15.  
- Conversely, there is a high abundance of red wines with alcohol content between 9 and 10.

## Scatterplot of citric acid vs fixed acidity

```python
plt.figure(figsize=(8,6))
plt.scatter(x=data['citric acid'], y=data['fixed acidity'], s=10, c='brown', alpha=0.5)
plt.title('Scatterplot of citric acid vs fixed acidity')
plt.xlabel('citric acid')
plt.ylabel('fixed acidity')
plt.show()
```

#### SCATTERPLOT REPORT Analysis of the scatter plot between the citric acid and fixed acidity  
- There is a strong linear relationship between these two because as the citric acid content increases the fixed acidity also increases  
- And there are very few outliers  
- Becauseof the less number of outliers it is not really significant  
- So there is a strong linear relation between the citric acid and fixed acidity
